```python
import pandas as pd
import seaborn as sns
```

```python
from pybaseball import batting_stats
```

```python
START= 2022
END= 2022
baseball_df= batting_stats(START, END, qual=200)
```

```python
# Compute the total runs scored by each team in each season
season_df = baseball_df.groupby(['Season', 'Team'])['R'].sum().reset_index()
```

```python
# Plot the relationship between season, team, and runs scored
sns.catplot(x='Season', y='R', hue='Team', data=season_df, kind='bar', height=3, aspect=2)
```

```python
season_df
```

```python
season_df= season_df.drop(season_df.index[0])
```

```python
season_df
```

```python
season_sort=season_df.sort_values(['R'])
```

```python
sns.catplot(x='Season', y='R', hue='Team', data=season_sort, kind='bar', height=3, aspect=2)
```

```python
import matplotlib.pyplot as plt
```

```python
corr = baseball_df['HR'].corr(baseball_df['R'])
```

```python
plt.scatter(baseball_df['HR'], baseball_df['R'])
plt.xlabel('Home Runs')
plt.ylabel('Points')
plt.title(f'Correlation: {corr:.2f}')
plt.show()
```

```python
baseball_df['Dist'] = baseball_df['1B'] + 2*baseball_df['2B'] + 3*baseball_df['3B'] + 4*baseball_df['HR']
mean_dist = baseball_df['Dist'].mean()
```

```python
plt.hist(baseball_df['Dist'], bins=50)
plt.axvline(mean_dist, color='r', linestyle='dashed', linewidth=2)
plt.xlabel('Distance to home plate')
plt.ylabel('Count')
plt.title(f'Mean distance: {mean_dist:.2f}')
plt.show()
```
